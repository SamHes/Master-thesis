```python
import json
```

```python
with open ('data/proposals/compound_proposals.json', 'r') as file:
    compound_proposals = json.load(file)['data']['proposals']
```

```python
markets = ['Ethereum', 'Arbitrum', 'Base', 'Optimism', 'Polygon'] # , 'Mainnet']?
tokens = ['AAVE', 'aAAVE', 'stkAAVE', 'COMP', 'cCOMP', 'USDC', 'USDT', 'WETH']
```

```python
compound_ids = []

for proposal in compound_proposals:
    matching_words = [word for word in markets if word.lower() in proposal['description'].lower()]
    
    if matching_words:
        id = proposal['id']
        compound_ids.append(id)
        
        print(f'Proposal {id} contains the following words {matching_words}')
```

```python
with open ('data/proposals/aave_snapshot_proposals.json', 'r', encoding='utf8') as file:
    aave_proposals = json.load(file)['data']['proposals']
```

```python
aave_ids = []

for proposal in aave_proposals:
    matching_words = [word for word in markets if word.lower() in proposal['title'].lower()]
    
    if matching_words:
        id = proposal['id']
        aave_ids.append(id)
        
        print(f'Proposal {id} contains the following words {matching_words}')
```

```python
print(f'Aave: total proposals: {len(aave_proposals)}, total with one of the markets: {len(aave_ids)}')
print(f'Compound: total proposals: {len(compound_proposals)}, total with one of the markets: {len(compound_ids)}')
```
